```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import math
import os

# Configuración para reproducibilidad
np.random.seed(42)

# Configuración de visualización
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 14

# 1. Generar datos sintéticos de serie temporal
def generar_serie_temporal(n_puntos=1000):
    """
    Genera una serie temporal sintética con tendencia, estacionalidad y ruido
    """
    # Componente de tendencia
    t = np.linspace(0, 10, n_puntos)
    tendencia = 0.5 * t

    # Componente estacional (múltiples frecuencias)
    estacionalidad1 = 2 * np.sin(2 * np.pi * 0.05 * t)  # Ciclo largo
    estacionalidad2 = 1 * np.sin(2 * np.pi * 0.2 * t)   # Ciclo medio
    estacionalidad3 = 0.5 * np.sin(2 * np.pi * 0.5 * t) # Ciclo corto

    # Ruido aleatorio
    ruido = 0.5 * np.random.normal(0, 1, n_puntos)

    # Serie temporal completa
    serie = tendencia + estacionalidad1 + estacionalidad2 + estacionalidad3 + ruido

    return serie, t

# 2. Preparar los datos para el entrenamiento
def crear_dataset(dataset, time_steps=1):
    """
    Convierte una serie temporal en un dataset para entrenamiento de RNN
    X contiene 'time_steps' valores consecutivos
    y contiene el siguiente valor después de X
    """
    X, y = [], []
    for i in range(len(dataset) - time_steps):
        X.append(dataset[i:(i + time_steps), 0])
        y.append(dataset[i + time_steps, 0])
    return np.array(X), np.array(y)

# 3. Crear callbacks para evitar sobreentrenamiento
def crear_callbacks(nombre_modelo='mejor_modelo_lstm'):
    """
    Crea callbacks para evitar sobreentrenamiento:
    - EarlyStopping: Detiene el entrenamiento cuando el rendimiento deja de mejorar
    - ModelCheckpoint: Guarda el mejor modelo según la métrica de validación
    - ReduceLROnPlateau: Reduce la tasa de aprendizaje cuando el rendimiento se estanca
    """
    # Asegurar que existe el directorio para guardar modelos
    if not os.path.exists('modelos'):
        os.makedirs('modelos')

    # Ruta para guardar el modelo
    ruta_modelo = os.path.join('modelos', f'{nombre_modelo}.h5')

    # Early Stopping: Detiene el entrenamiento cuando la pérdida de validación deja de mejorar
    early_stopping = EarlyStopping(
        monitor='val_loss',      # Métrica a monitorear
        patience=10,             # Número de épocas sin mejora antes de detenerse
        verbose=1,               # Mostrar mensajes
        restore_best_weights=True # Restaurar los mejores pesos encontrados
    )

    # Model Checkpoint: Guarda el mejor modelo según la pérdida de validación
    checkpoint = ModelCheckpoint(
        filepath=ruta_modelo,    # Ruta donde guardar el modelo
        monitor='val_loss',      # Métrica a monitorear
        save_best_only=True,     # Guardar solo el mejor modelo
        verbose=1,               # Mostrar mensajes
        mode='min'               # Queremos minimizar la pérdida
    )

    # Reduce LR On Plateau: Reduce la tasa de aprendizaje cuando la métrica se estanca
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',      # Métrica a monitorear
        factor=0.2,              # Factor por el que se reduce la tasa de aprendizaje
        patience=5,              # Número de épocas sin mejora antes de reducir la tasa
        min_lr=1e-6,             # Tasa de aprendizaje mínima
        verbose=1                # Mostrar mensajes
    )

    return [early_stopping, checkpoint, reduce_lr]

# 4. Visualizar la historia del entrenamiento
def visualizar_historia_entrenamiento(history):
    """
    Visualiza la evolución de la pérdida durante el entrenamiento y la tasa de aprendizaje
    """
    plt.figure(figsize=(16, 6))

    # Gráfico de pérdida
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Entrenamiento')
    plt.plot(history.history['val_loss'], label='Validación')
    plt.title('Evolución de la Pérdida Durante el Entrenamiento')
    plt.xlabel('Época')
    plt.ylabel('Error Cuadrático Medio')
    plt.legend()
    plt.grid(True)

    # Gráfico de tasa de aprendizaje (si está disponible)
    if 'lr' in history.history:
        plt.subplot(1, 2, 2)
        plt.plot(history.history['lr'])
        plt.title('Evolución de la Tasa de Aprendizaje')
        plt.xlabel('Época')
        plt.ylabel('Tasa de Aprendizaje')
        plt.yscale('log')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

# 5. Crear y entrenar el modelo LSTM
def crear_y_entrenar_modelo_lstm(X_train, y_train, time_steps, epochs=100, batch_size=32):
    """
    Crea y entrena un modelo LSTM con callbacks para evitar sobreentrenamiento
    """
    # Crear el modelo
    modelo = Sequential()

    # Primera capa LSTM con 50 unidades y regularización dropout
    modelo.add(LSTM(units=50, return_sequences=True,
                   input_shape=(time_steps, 1)))
    modelo.add(Dropout(0.2))

    # Segunda capa LSTM
    modelo.add(LSTM(units=50, return_sequences=False))
    modelo.add(Dropout(0.2))

    # Capa de salida
    modelo.add(Dense(units=1))

    # Compilar el modelo
    modelo.compile(optimizer='adam', loss='mean_squared_error')

    # Mostrar resumen del modelo
    modelo.summary()

    # Crear callbacks
    callbacks = crear_callbacks('lstm_serie_temporal')

    # Entrenar el modelo con validación
    history = modelo.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_split=0.2,  # Usar 20% de los datos de entrenamiento para validación
        callbacks=callbacks,   # Usar los callbacks definidos
        verbose=1
    )

    return modelo, history

# 6. Visualizar predicciones
def visualizar_predicciones(serie, tiempo, train_size, time_steps, y_train_pred, y_test_pred):
    """
    Visualiza las predicciones del modelo vs los datos reales
    """
    # Preparar los datos para la visualización
    train_plot = np.empty_like(serie)
    train_plot[:] = np.nan
    train_plot[time_steps:train_size] = y_train_pred

    test_plot = np.empty_like(serie)
    test_plot[:] = np.nan
    test_plot[train_size+time_steps:len(serie)] = y_test_pred

    # Crear gráfico con datos originales y predicciones
    plt.figure(figsize=(16, 8))
    plt.plot(tiempo, serie, label='Serie Original', alpha=0.7)
    plt.plot(tiempo, train_plot, label='Predicción en Entrenamiento', color='red')
    plt.plot(tiempo, test_plot, label='Predicción en Prueba', color='green')

    # Añadir línea vertical para separar entrenamiento y prueba
    plt.axvline(x=tiempo[train_size], color='k', linestyle='--', alpha=0.3)
    plt.text(tiempo[train_size+50], max(serie), 'Entrenamiento | Prueba',
             fontsize=14, verticalalignment='top')

    plt.title('Predicción de Serie Temporal con LSTM')
    plt.xlabel('Tiempo')
    plt.ylabel('Valor')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 7. Función principal
def main():
    # Generar datos ficticios
    print("Generando serie temporal sintética...")
    serie, tiempo = generar_serie_temporal(1500)

    # Visualizar la serie temporal generada
    plt.figure(figsize=(16, 6))
    plt.plot(tiempo, serie, label='Serie Temporal')
    plt.title('Serie Temporal Sintética con Tendencia, Estacionalidad y Ruido')
    plt.xlabel('Tiempo')
    plt.ylabel('Valor')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Dividir en conjuntos de entrenamiento y prueba
    train_size = int(len(serie) * 0.8)
    test_size = len(serie) - train_size
    train, test = serie[0:train_size], serie[train_size:len(serie)]
    print(f"Tamaño del conjunto de entrenamiento: {train_size}")
    print(f"Tamaño del conjunto de prueba: {test_size}")

    # Normalizar los datos
    scaler = MinMaxScaler(feature_range=(0, 1))
    train_scaled = scaler.fit_transform(train.reshape(-1, 1))
    test_scaled = scaler.transform(test.reshape(-1, 1))

    # Definir el número de pasos de tiempo (ventana de observación)
    time_steps = 60

    # Crear datasets con estructura para RNN
    X_train, y_train = crear_dataset(train_scaled, time_steps)
    X_test, y_test = crear_dataset(test_scaled, time_steps)

    # Reshape para formato de entrada de LSTM [muestras, time_steps, características]
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    print(f"Forma de X_train: {X_train.shape}")
    print(f"Forma de y_train: {y_train.shape}")

    # Crear y entrenar el modelo LSTM
    print("Entrenando el modelo LSTM con callbacks...")
    modelo, history = crear_y_entrenar_modelo_lstm(
        X_train, y_train,
        time_steps=time_steps,
        epochs=100,  # Máximo número de épocas (early stopping puede detenerlo antes)
        batch_size=32
    )

    # Visualizar la evolución del entrenamiento
    visualizar_historia_entrenamiento(history)

    # Hacer predicciones
    train_predict = modelo.predict(X_train)
    test_predict = modelo.predict(X_test)

    # Invertir la normalización para obtener valores originales
    train_predict = scaler.inverse_transform(train_predict)
    y_train_inv = scaler.inverse_transform(y_train.reshape(-1, 1))
    test_predict = scaler.inverse_transform(test_predict)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Calcular RMSE (Root Mean Squared Error)
    train_rmse = math.sqrt(mean_squared_error(y_train_inv, train_predict))
    test_rmse = math.sqrt(mean_squared_error(y_test_inv, test_predict))
    print(f'RMSE en entrenamiento: {train_rmse:.4f}')
    print(f'RMSE en prueba: {test_rmse:.4f}')

    # Visualizar las predicciones
    visualizar_predicciones(
        serie, tiempo, train_size, time_steps,
        train_predict.flatten(), test_predict.flatten()
    )

    print("Análisis completo de la serie temporal con LSTM")

# Ejecutar el programa principal
if __name__ == "__main__":
    main()
```
